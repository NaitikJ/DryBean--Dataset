## Dry Bean Dataset Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Reading the data

```python
df = pd.read_csv('Dry_Bean_Dataset.csv')
```

#### Exploring the data

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
null = df.isnull().sum().to_frame()
null.columns = ['No.of Null Values']
null
```

```python
skew = df.skew().to_frame()
skew.columns = ['Skewness']
skew
```

```python
kurt = df.kurt().to_frame()
kurt.columns = ['Kurtosis']
kurt
```

```python
df['Class'].unique()
```

```python
plt.figure(figsize=(12,6))
plt.title('Class Count')
sns.set_style('white')
sns.countplot(x='Class',data=df)
```

### This shows that the dataset is unbalanced having unqual records for the classes.

# Visualisation

```python
plt.figure(figsize=(10,6))
sns.set_style('white')
sns.boxplot(x="Class",y="roundness",data=df)
```

### Seker Class is more round compared to other Classes

```python
plt.figure(figsize=(10,6))
sns.set_style('white')
sns.boxplot(x="Class",y="Area",data=df)
```

#### Bombay class has largest area, i.e largest in size

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Class", y="Eccentricity", data=df,palette='rainbow')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df['Area'],y=df['roundness'],hue=df['Class'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x= df['ShapeFactor3'] ,y =df['Area'],hue=df['Class'],alpha=0.8)
plt.xlabel('ShapeFactor3')
plt.ylabel('Area')
plt.legend()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= df['AspectRation'],y = df['ShapeFactor1'],hue=df['Class'],alpha=0.8)
plt.xlabel('AspectRation')
plt.ylabel('ShapeFactor1')
plt.legend()
```

```python
corr = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr,cmap='Blues')
```

```python
cor = corr[corr <0.5]
cor
```

Selecting the parameters which are not highly correlated to each other.(corr < 0.5).If one paramter is correlated with other then we take only one of them for analysis.
eg.ShapeFactor1 is highly corelated with shapeFactor2,shapeFactor3,shapeFactor4, so we pick only shapeFactor1
